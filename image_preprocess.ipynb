# Data Preparation: Extract- Transform-Load

在TensorFlow中读取数据一般有3种方法：

-  feeding:使用placeholder读内存中的数据(np.narray):  
  会提供更高的灵活性。
  使用feed_dict的大多数实例不可以进行合适的扩展。然而，在只有一个GPU的实例中，这种差异是微不足道的。
  

- QueueRunner:使用queue读硬盘中的数据:   
  基于python的queue_runner（受限于python的多线程低性能）  
  unfortunately, it was somewhat (extremely) difficult to get the queues working correctly.
  
  
- Dataset API  
  使用C++的多线程机制的开销更低。  
  该API在tensorFlow 1.2中作为contrib的一部分被添加进去，并在后续版本会移至core包中。  
  The two main abstractions introduced in the tf.data API are: 
      tf.data.Dataset 
      tf.data.Iterator  
  ResNet example 会训练CIFAR-10,它展示了如何使用Dataset API以及tf.estimator.Estimator。

## Binary version
The binary version contains the files data_batch_1.bin, data_batch_2.bin, ..., data_batch_5.bin, as well as test_batch.bin. Each of these files is formatted as follows:
<1 x label><3072 x pixel>
...
<1 x label><3072 x pixel>
In other words, the first byte is the label of the first image, which is a number in the range 0-9. The next 3072 bytes are the values of the pixels of the image. The first 1024 bytes are the red channel values, the next 1024 the green, and the final 1024 the blue. The values are stored in row-major order, so the first 32 bytes are the red channel values of the first row of the image. 

Each file contains 10000 such 3073-byte "rows" of images, although there is nothing delimiting the rows. Therefore each file should be exactly 30730000 bytes long. 

There is another file, called batches.meta.txt. This is an ASCII file that maps numeric labels in the range 0-9 to meaningful class names. It is merely a list of the 10 class names, one per row. The class name on row i corresponds to numeric label i.M

```python
data_dir ='D:\wangfeicheng\Tensorflow\cifar10-tensorflow\cifar-10-binary\cifar-10-batches-bin'
filename = os.path.join(data_dir, 'data_batch_%d.bin' % 1)

record_bytes = 32*32*3+1
image_num =10

print(filename)
with open(filename,'rb') as f:
    data_btyes = f.read(record_bytes*image_num)
    print(type(data_btyes))
    data = np.fromstring(data_btyes, dtype=np.ubyte)
    
    print(type(data),data.shape)
    labels = []
    images = []
    for i in range(image_num):
        begin_byte= i*record_bytes
        label = data[begin_byte]
        image_bytes = data[begin_byte+1:begin_byte+record_bytes]
        image = image_bytes.reshape( (3,32,32)).transpose([1,2,0])
        labels.append(label)
        images.append(image)
        
        test_path = r'D:\wangfeicheng\Tensorflow\tensorflow-learning\cifar10\test_{}.png'.format(i)
        if os.path.exists(test_path):
            img = open(test_path,'w')
            img.close()

        ima = Image.fromarray(image)
        ima.save(test_path)
        
        with open(test_path+'.binary','wb') as f:
            f.write(image_bytes)

label_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
label_dict = {}
for key, value in enumerate(label_names):
    label_dict[key] = value
    
print(len(images))
plt.figure(figsize = (16,16))
for i in range(image_num):
    plt.subplot(2,5,i+1)
    plt.imshow(images[i])
    plt.title(label_dict[labels[i]])
plt.show()
```

## 数据类型转化  
	convert_image_dtype(image,dtype,saturate=False,name=None)
		作用：把图片元素类型，转成想要的类型，返回转换后的图片,注意，要是转成了float类型之后，像素值会在 [0,1)这个范围内。 
		参数: 
		image: 图像 
		dtype: 待转换类型 
		saturate: If True, clip the input before casting (if necessary). 
		name: 可选操作名

```python

print(file_decoded)
image_float = tf.image.convert_image_dtype(file_decoded,dtype=tf.float32)
image = image_float.eval()
print(image[:,:,0])
plt.imshow(image)
```
